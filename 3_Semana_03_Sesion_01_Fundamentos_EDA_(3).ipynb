#### Explicación del código de detección de valores faltantes

A continuación, se explica línea por línea el código utilizado para detectar valores faltantes en el DataFrame `datos_ingenieria`:

```python
# ==========================================
# PASO 2: VALORES FALTANTES

print("\n--- 2. DETECCIÓN DE VALORES FALTANTES ---")
faltantes = datos_ingenieria.isna().sum()
print(faltantes[faltantes > 0])

```

**Explicación Línea por Línea:**

*   `print("\n--- 2. DETECCIÓN DE VALORES FALTANTES ---")`: Esta línea simplemente imprime un encabezado en la consola para indicar el inicio de la sección de detección de valores faltantes.
*   `faltantes = datos_ingenieria.isna().sum()`:
    *   `datos_ingenieria.isna()`: Este método de pandas devuelve un DataFrame del mismo tamaño que `datos_ingenieria` donde cada celda es `True` si el valor original era `NaN` (Not a Number, un valor faltante) y `False` en caso contrario.
    *   `.sum()`: Al aplicar `.sum()` a este DataFrame booleano, Python trata `True` como 1 y `False` como 0. Por lo tanto, `sum()` calcula la cantidad total de valores `True` (es decir, valores faltantes) para cada columna. El resultado se guarda en la variable `faltantes`, que es una Serie de pandas.
*   `print(faltantes[faltantes > 0])`: Esta línea imprime solo las columnas que tienen al menos un valor faltante.
    *   `faltantes > 0`: Esto crea una Serie booleana donde `True` indica que la columna correspondiente en `faltantes` tiene más de 0 valores faltantes.
    *   `faltantes[...]`: Al usar esta Serie booleana para indexar `faltantes`, solo se seleccionan e imprimen las entradas donde la condición es `True`, mostrando así únicamente las columnas con valores faltantes y el conteo de los mismos.
*   `# Interpretación: Saber cuántos faltan y en qué variables...`: Este es un comentario que proporciona una guía sobre cómo interpretar los resultados obtenidos, sugiriendo la importancia de esta información para decidir si imputar datos o descartar registros.

#### **Celda 5: Código Python (Paso 4)**

````
# ==========================================
# PASO 4: FORMA DE LA DISTRIBUCIÓN
# ==========================================
print("\n--- 4. ASIMETRÍA (SKEWNESS) Y CURTOSIS ---")

# Calculamos asimetría y curtosis solo para variables numéricas, omitiendo los NaN
resistencia_limpia = datos_ingenieria['Resistencia_MPa'].dropna()

asimetria = skew(resistencia_limpia)
curtosis_val = kurtosis(resistencia_limpia)

print(f"Asimetría de la Resistencia: {asimetria:.2f}")
print(f"Curtosis de la Resistencia: {curtosis_val:.2f}")

# Interpretación Ingenieril:
# - Asimetría cercana a 0 indica distribución simétrica.
# - Curtosis > 0 indica colas pesadas (mayor probabilidad de valores extremos o fallos inesperados en el material).
````

#### Explicación del código para la Forma de la Distribución

A continuación, se explica línea por línea el código utilizado para calcular la asimetría y curtosis de la distribución de `Resistencia_MPa`:

```python
# ==========================================
# PASO 4: FORMA DE LA DISTRIBUCIÓN
# ==========================================
print("\n--- 4. ASIMETRÍA (SKEWNESS) Y CURTOSIS ---")

# Calculamos asimetría y curtosis solo para variables numéricas, omitiendo los NaN
resistencia_limpia = datos_ingenieria['Resistencia_MPa'].dropna()

asimetria = skew(resistencia_limpia)
curtosis_val = kurtosis(resistencia_limpia)

print(f"Asimetría de la Resistencia: {asimetria:.2f}")
print(f"Curtosis de la Resistencia: {curtosis_val:.2f}")

# Interpretación Ingenieril:
# - Asimetría cercana a 0 indica distribución simétrica.
# - Curtosis > 0 indica colas pesadas (mayor probabilidad de valores extremos o fallos inesperados en el material).
```

**Explicación Línea por Línea:**

*   `print("\n--- 4. ASIMETRÍA (SKEWNESS) Y CURTOSIS ---")`: Imprime un encabezado para indicar el inicio de esta sección del análisis.
*   `resistencia_limpia = datos_ingenieria['Resistencia_MPa'].dropna()`:
    *   `datos_ingenieria['Resistencia_MPa']`: Selecciona la columna 'Resistencia_MPa' del DataFrame `datos_ingenieria`.
    *   `.dropna()`: Este método elimina cualquier valor `NaN` (valores faltantes) de la serie resultante. Es crucial para que las funciones `skew` y `kurtosis` puedan operar correctamente, ya que no manejan `NaN` directamente.
    *   El resultado, una serie sin valores faltantes, se guarda en la variable `resistencia_limpia`.
*   `asimetria = skew(resistencia_limpia)`: Calcula la asimetría (skewness) de la distribución de los datos en `resistencia_limpia`. La asimetría mide la simetría de la distribución. Un valor cercano a 0 indica una distribución casi simétrica.
*   `curtosis_val = kurtosis(resistencia_limpia)`: Calcula la curtosis de la distribución de los datos en `resistencia_limpia`. La curtosis mide la "puntiagudez" de la distribución y la "pesadez" de sus colas en comparación con una distribución normal. Un valor positivo indica colas más pesadas y un pico más pronunciado (leptocúrtica), mientras que un valor negativo indica colas más ligeras y un pico más plano (platocúrtica).
*   `print(f"Asimetría de la Resistencia: {asimetria:.2f}")`: Imprime el valor calculado de la asimetría, formateado a dos decimales.
*   `print(f"Curtosis de la Resistencia: {curtosis_val:.2f}")`: Imprime el valor calculado de la curtosis, formateado a dos decimales.
*   `# Interpretación Ingenieril:` (y las líneas siguientes): Estos son comentarios que proporcionan una guía sobre cómo interpretar los resultados de asimetría y curtosis en un contexto de ingeniería, destacando lo que implican valores específicos.
