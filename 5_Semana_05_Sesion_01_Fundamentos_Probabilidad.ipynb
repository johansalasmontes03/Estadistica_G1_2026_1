## Fundamentos de Probabilidad para Ingeniería

### Contexto en Ingeniería
En ingeniería, la probabilidad es la herramienta fundamental que nos permite cuantificar y modelar la incertidumbre. Ya sea en la ingeniería civil (estimando la falla estructural), agrícola (prediciendo el rendimiento de cosechas bajo condiciones variables) o agroindustrial (controlando la calidad de la producción), entender la probabilidad nos capacita para tomar decisiones más robustas y gestionar riesgos de manera efectiva.

### Conceptos Fundamentales
Para hablar de probabilidad, necesitamos definir algunos elementos clave:

*   **Experimento Aleatorio:** Cualquier proceso cuyo resultado no se puede predecir con certeza, pero cuyos posibles resultados son conocidos de antemano. Ejemplos incluyen el lanzamiento de un dado, la medición de la resistencia de un material o la inspección de un lote de productos.
*   **Espacio Muestral ($\Omega$):** El conjunto de todos los posibles resultados de un experimento aleatorio. Por ejemplo, al lanzar un dado, $\Omega = \{1, 2, 3, 4, 5, 6\}$.
*   **Evento ($A$):** Cualquier subconjunto del espacio muestral. Un evento es un resultado o una colección de resultados específicos. Por ejemplo, "obtener un número par" al lanzar un dado es un evento: $A = \{2, 4, 6\}$.

### Ecuaciones de Probabilidad
Existen dos enfoques principales para calcular la probabilidad de un evento:

*   **Probabilidad Teórica (Regla de Laplace):** Cuando todos los resultados en el espacio muestral son igualmente probables, la probabilidad de un evento $A$ se calcula como:
    $$
P(A) = \frac{\text{Número de casos favorables a } A}{\text{Número total de casos posibles}}
    $$

*   **Probabilidad Empírica (Frecuencia Relativa):** Basada en la observación de datos reales o la repetición de un experimento un gran número de veces. Se calcula como:
    $$
P(A) = \frac{\text{Frecuencia de ocurrencia de } A}{\text{Número total de observaciones}}
    $$

### Transición Práctica
Ahora, llevaremos estos conceptos abstractos a la práctica, utilizando Python para simular y calcular probabilidades empíricas a partir de datos.

````
import pandas as pd
import numpy as np

# Simulamos el registro histórico de inspección de 500 vigas de concreto (Ing. Civil)
# o lotes de producción (Ing. Agroindustrial).
# Resultados posibles (Espacio Muestral): "Aprobado", "Fallo_Menor", "Fallo_Critico"

np.random.seed(42)
espacio_muestral = ['Aprobado', 'Fallo_Menor', 'Fallo_Critico']
# Asignamos una probabilidad teórica subyacente (80% éxito, 15% fallo menor, 5% fallo crítico)
inspecciones = np.random.choice(espacio_muestral, size=500, p=[0.80, 0.15, 0.05])

df_calidad = pd.DataFrame({'Estado_Inspeccion': inspecciones})

# ==========================================
# CÁLCULO DE PROBABILIDAD EMPÍRICA
# ==========================================
print("--- FRECUENCIAS ABSOLUTAS ---")
frecuencias = df_calidad['Estado_Inspeccion'].value_counts()
print(frecuencias)

print("\n--- PROBABILIDAD EMPÍRICA (Frecuencia Relativa) ---")
prob_empirica = df_calidad['Estado_Inspeccion'].value_counts(normalize=True)
print(prob_empirica.round(4))

# Interpretación Ingenieril:
# El "Espacio Muestral" son los 3 estados posibles.
# Un "Evento" puede ser que una viga tenga un 'Fallo_Critico'.
# La Probabilidad Empírica de 'Fallo_Critico' calculada con nuestros datos es cercana a 0.05 (5%),
# calculada simplemente dividiendo los casos favorables entre el total de inspecciones.
````

## Probabilidad Condicional e Independencia en Ingeniería

### Introducción Práctica
Hasta ahora, hemos calculado la probabilidad empírica de eventos individuales. Pero en ingeniería, a menudo nos enfrentamos a situaciones donde la ocurrencia de un evento podría influir en la probabilidad de otro. Por ejemplo, ¿la probabilidad de un fallo crítico en una viga de concreto (ingeniería civil) o un lote de productos (ingeniería agroindustrial) cambia si sabemos que el material proviene de un proveedor específico? Responder a estas preguntas nos lleva al concepto de probabilidad condicional.

### Conceptos Clave

*   **Probabilidad Condicional:** Es la probabilidad de que ocurra un evento $A$, dado que otro evento $B$ ya ha ocurrido. Nos permite actualizar nuestra creencia o medida de incertidumbre sobre $A$ basándonos en nueva información ($B$).

*   **Independencia Estadística:** Dos eventos $A$ y $B$ son estadísticamente independientes si la ocurrencia de uno no afecta la probabilidad de ocurrencia del otro. En un contexto de control de calidad, si la probabilidad de un defecto es la misma independientemente del proveedor, entonces el defecto y el proveedor son eventos independientes.

### Ecuaciones en LaTeX

*   **Probabilidad Condicional:** La probabilidad de $A$ dado $B$ se define como:
    $$
P(A|B) = \frac{P(A \cap B)}{P(B)}
    $$
    donde $P(A \cap B)$ es la probabilidad de que ambos eventos $A$ y $B$ ocurran, y $P(B)$ es la probabilidad de que ocurra el evento $B$ (asumiendo $P(B) > 0$).

*   **Condición Matemática de Independencia:** Dos eventos $A$ y $B$ son independientes si y solo si:
    $$P(A|B) = P(A)$$
    o, equivalentemente,
    $$P(A \cap B) = P(A) \cdot P(B)$$

### Transición al Código (Tablas de Contingencia)
A continuación, utilizaremos la librería Pandas de Python para construir tablas de contingencia cruzadas a partir de nuestros datos simulados, lo que nos permitirá calcular estas probabilidades condicionales y evaluar la independencia entre los estados de inspección y los proveedores.

## **Celda 4: Código Python (Probabilidad Condicional e Independencia)**

````
# Ahora, supongamos que estas inspecciones provienen de dos proveedores distintos
proveedores = np.random.choice(['Proveedor_A', 'Proveedor_B'], size=500, p=[0.6, 0.4])
df_calidad['Proveedor'] = proveedores

# Tabla de contingencia cruzada
tabla_cruzada = pd.crosstab(df_calidad['Proveedor'], df_calidad['Estado_Inspeccion'], margins=True)

print("--- TABLA DE CONTINGENCIA ---")
print(tabla_cruzada)

# Cálculo de Probabilidad Condicional
# P(Fallo_Critico | Proveedor_B) = Casos de Fallo Crítico en Prov_B / Total de Prov_B
total_prov_b = tabla_cruzada.loc['Proveedor_B', 'All']
fallos_prov_b = tabla_cruzada.loc['Proveedor_B', 'Fallo_Critico']

prob_condicional = fallos_prov_b / total_prov_b
print(f"\nProbabilidad de Fallo Crítico DADO que es del Proveedor B: {prob_condicional:.4f}")

# Interpretación: Si la probabilidad condicional es igual a la probabilidad marginal (general),
# decimos que los eventos son INDEPENDIENTES. En la práctica ingenieril, si un proveedor
# tiene una tasa de fallos mucho mayor, la calidad "depende" del proveedor.
````

### Explicación del Código Python (Probabilidad Condicional e Independencia):

*   `proveedores = np.random.choice(['Proveedor_A', 'Proveedor_B'], size=500, p=[0.6, 0.4])`: Simula los proveedores para cada inspección, asignando aleatoriamente 'Proveedor_A' al 60% de las inspecciones y 'Proveedor_B' al 40%.
*   `df_calidad['Proveedor'] = proveedores`: Añade una nueva columna 'Proveedor' al DataFrame `df_calidad` con los datos de proveedor simulados.
*   `tabla_cruzada = pd.crosstab(df_calidad['Proveedor'], df_calidad['Estado_Inspeccion'], margins=True)`: Crea una tabla de contingencia (tabla cruzada) usando Pandas, que muestra la frecuencia conjunta de los estados de inspección y los proveedores, incluyendo los totales marginales (`margins=True`).
*   `print('--- TABLA DE CONTINGENCIA ---')`: Imprime un encabezado para la sección de la tabla de contingencia.
*   `print(tabla_cruzada)`: Muestra la tabla de contingencia.
*   `total_prov_b = tabla_cruzada.loc['Proveedor_B', 'All']`: Extrae el número total de inspecciones realizadas por el 'Proveedor_B' de la tabla de contingencia.
*   `fallos_prov_b = tabla_cruzada.loc['Proveedor_B', 'Fallo_Critico']`: Extrae el número de 'Fallo_Critico' reportados para el 'Proveedor_B' de la tabla de contingencia.
*   `prob_condicional = fallos_prov_b / total_prov_b`: Calcula la probabilidad condicional de un 'Fallo_Critico' dado que el producto proviene del 'Proveedor_B' dividiendo los fallos críticos del Proveedor B por el total de productos del Proveedor B.
*   `print(f'\nProbabilidad de Fallo Crítico DADO que es del Proveedor B: {prob_condicional:.4f}')`: Imprime la probabilidad condicional calculada, formateada a cuatro decimales.
